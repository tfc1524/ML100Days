```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv', header=None)
label = pd.read_csv('trainLabels.csv', header=None)
test = pd.read_csv('test.csv', header=None)
```

```python
print(train.shape)
print(label.shape)
print(test.shape)
```

```python
# baseline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, label, random_state=0)
forest = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0)

forest.fit(X_train, y_train.values.ravel())
y_pred = forest.predict(X_test)

print('CV:', np.mean(cross_val_score(forest, X_train, y_train.values.ravel())))
print('Accuracy:', accuracy_score(y_pred, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

logic = LogisticRegression()
logic.fit(X_train, y_train.values.ravel())
y_pred = logic.predict(X_test)

print('CV:', np.mean(cross_val_score(logic, X_train, y_train.values.ravel())))
print('Accuracy:', accuracy_score(y_pred, y_test))
```

特徵縮放 全部變數都是常態分佈所以使用標準化

```python
from sklearn.preprocessing import StandardScaler

stdscaler = StandardScaler()
std_train = stdscaler.fit_transform(train)

X_train, X_test, y_train, y_test = train_test_split(std_train, label, random_state=0)
forest = RandomForestClassifier(n_estimators=1000, max_depth=15, random_state=0)

forest.fit(X_train, y_train.values.ravel())
y_pred = forest.predict(X_test)

print('CV:', np.mean(cross_val_score(forest, X_train, y_train.values.ravel())))
print('Accuracy:', accuracy_score(y_pred, y_test))
```

```python
logic = LogisticRegression()
logic.fit(X_train, y_train.values.ravel())
y_pred = logic.predict(X_test)

print('CV:', np.mean(cross_val_score(logic, X_train, y_train.values.ravel())))
print('Accuracy:', accuracy_score(y_pred, y_test))
```

```python
# 隨機森林調參數
from sklearn.model_selection import KFold, GridSearchCV, RandomizedSearchCV

forest = RandomForestClassifier()
distribution = {'n_estimators': range(0, 1001), 'max_depth': range(0, 25)}

random_search = RandomizedSearchCV(forest, distribution, random_state=0)
random_result = random_search.fit(X_train, y_train.values.ravel())
print("Best Accuracy: %f using %s" % (random_result.best_score_, random_result.best_params_))

# 使用最佳參數重新建立模型
forest_bestparam = RandomForestClassifier(max_depth=random_result.best_params_['max_depth'],
                                          n_estimators=random_result.best_params_['n_estimators'])

# 訓練模型
forest_bestparam.fit(X_train, y_train.values.ravel())

# 預測測試集
y_pred = forest_bestparam.predict(X_test)

print('CV:', np.mean(cross_val_score(forest_bestparam, X_train, y_train.values.ravel())))
print('Random Forest Accuracy',cross_val_score(forest_bestparam, X_train, y_train.values.ravel(), cv=10).mean())
print('Accuracy:', accuracy_score(y_pred, y_test))
```

```python
std_test = stdscaler.fit_transform(test)

pred  = forest_bestparam.predict(std_test)
forest_best_pred = pd.DataFrame(pred)

forest_best_pred.index += 1

forest_best_pred.columns = ['Solution']
forest_best_pred['Id'] = np.arange(1,forest_best_pred.shape[0]+1)
forest_best_pred = forest_best_pred[['Id', 'Solution']]

forest_best_pred.to_csv('Submission_forest.csv',index=False)
```
