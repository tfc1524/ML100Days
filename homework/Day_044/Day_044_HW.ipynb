```python
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# 讀取鳶尾花資料集
iris = datasets.load_iris()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=0)

# 建立模型 (使用 20 顆樹，每棵樹的最大深度為 4)
clf = RandomForestClassifier(n_estimators=1000, max_depth=20)

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)

fi = pd.DataFrame(clf.feature_importances_, index=iris.feature_names, columns=['importance'])
fi = fi.sort_values(by='importance', ascending=False)
fi
```

```python
# 讀取鳶尾花資料集
breast_cancer = datasets.load_breast_cancer()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.25, random_state=0)

# 建立模型 (使用 20 顆樹，每棵樹的最大深度為 4)
clf = RandomForestClassifier(n_estimators=1000, max_depth=20)

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)

fi = pd.DataFrame(clf.feature_importances_, index=breast_cancer.feature_names, columns=['importance'])
fi = fi.sort_values(by='importance', ascending=False)
fi
```

```python
# 讀取鳶尾花資料集
wine = datasets.load_wine()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=0)

# 建立模型 (使用 20 顆樹，每棵樹的最大深度為 4)
reg = RandomForestRegressor(n_estimators=1000, max_depth=20, criterion='mse')

# 訓練模型
reg.fit(x_train, y_train)

# 預測測試集
y_pred = reg.predict(x_test)

acc = metrics.mean_squared_error(y_test, y_pred)
print("MSE: ", acc)

fi = pd.DataFrame(reg.feature_importances_, index=wine.feature_names, columns=['importance'])
fi = fi.sort_values(by='importance', ascending=False)
fi
```
