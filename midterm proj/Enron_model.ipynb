```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train_data.csv', index_col=0)
train = train.reset_index()
y = train['poi']
y
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, GridSearchCV, RandomizedSearchCV
from sklearn.svm import SVC
import eli5
from eli5.sklearn import PermutationImportance

file_list = ['丟棄過低、中離散化、高補0+全MM.csv', '丟棄過低、中離散化、高補0標準化.csv',
             '丟棄過低+全補0不動.csv', '全補0+bool+log+MM.csv', '全離群化.csv']
score = dict()
```

```python
X = pd.read_csv('丟棄過低、中離散化、高補0+全MM.csv', index_col=0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

print('丟棄過低、中離散化、高補0+全MM.csv')
print('CV: ', np.mean(cross_val_score(rfc, X_train, y_train)))
print('Accuracy: ', metrics.accuracy_score(y_pred, y_test))

score['丟棄過低、中離散化、高補0+全MM.csv'] = metrics.accuracy_score(y_pred, y_test)
```

```python
test = pd.read_csv('test_丟棄過低、中離散化、高補0+MM.csv', index_col=0)
test.head(5)
```

```python
name = test['name']
test.drop('name', axis=1, inplace=True)
```

```python
test_pred = rfc.predict_proba(test)[:,1]
sub = pd.DataFrame({'name': name, 'poi': test_pred})
sub
```

```python
sub.to_csv('basic_submission.csv', index=False)
```
