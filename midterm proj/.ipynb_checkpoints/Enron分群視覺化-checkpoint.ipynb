```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

視覺化heatmap裡面最熱的三個變數、兩個通信ratio

```python
df = pd.read_csv('train_data.csv')
y = df['poi']
df.drop(['name', 'email_address'], axis=1, inplace=True)
df.drop(['loan_advances'], axis=1, inplace=True)
print(df.shape)
df.head(5)
```

```python
fin_features = ['salary', 'deferral_payments', 'total_payments', 'bonus', 'restricted_stock_deferred', 'deferred_income', 'director_fees',
                'total_stock_value', 'expenses', 'exercised_stock_options', 'other', 'long_term_incentive', 'restricted_stock']
email_features = ['to_messages', 'from_poi_to_this_person', 'from_messages', 
                  'from_this_person_to_poi', 'shared_receipt_with_poi']
```

```python
df['poi'].value_counts()
```

```python
df.info()
```

```python
for i in range(len(df.columns)):
    if df.columns[i] != 'poi':
        plt.figure(figsize=(10,5))
        plt.title(df.columns[i])
        sns.histplot(df[df.columns[i]], kde=True)
plt.show()
```

```python
for i in range(len(df.columns)):
    if df.columns[i] != 'poi':
        df[df.columns[i]] = np.log1p(abs(df[df.columns[i]]))
        plt.figure(figsize=(10,5))
        plt.title(df.columns[i])
        sns.histplot(df[df.columns[i]], kde=True)
plt.show()
```

```python
normal_cols = ['bonus', 'deferral_payments', 'deferred_income', 'total_stock_value', 'to_messages', 
               'exercised_stock_options', 'from_poi_to_this_person', '']
```

```python
# 對數轉換後的boxplot
for i in range(len(df.columns)):
    if df.columns[i] != 'poi':
        plt.figure(figsize=(10,5))
        plt.title(df.columns[i])
        sns.boxplot(data=df[df.columns[i]])
plt.show()
```

```python
# 基於對資料的了解，我們可以知道空值全都為0
df = df.fillna(0)
```

```python
corrs = df.corr()

plt.figure(figsize = (20, 20))
# 繪製相關係數 (correlations) 的 Heatmap
sns.heatmap(corrs, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap');

# poi相關度較高(abs > 0.3)的欄位：'exercised_stock_options', 'salary', 'total_stock_value'
```
